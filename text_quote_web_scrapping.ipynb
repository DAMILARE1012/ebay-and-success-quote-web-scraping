```python
# Imported libraries...
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import requests
```

```python
# Web URL to be scrapped...
quote_toscrape_web_page = "http://quotes.toscrape.com/"

headers = {
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7", 
    "Accept-Encoding": "gzip, deflate, br", 
    "Accept-Language": "en-US,en;q=0.9", 
    "Dnt": "1", 
    "Host": "httpbin.org", 
    "Sec-Ch-Ua": "\"Not_A Brand\";v=\"8\", \"Chromium\";v=\"120\", \"Google Chrome\";v=\"120\"", 
    "Sec-Ch-Ua-Mobile": "?0", 
    "Sec-Ch-Ua-Platform": "\"Windows\"", 
    "Sec-Fetch-Dest": "document", 
    "Sec-Fetch-Mode": "navigate", 
    "Sec-Fetch-Site": "none", 
    "Sec-Fetch-User": "?1", 
    "Upgrade-Insecure-Requests": "1", 
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36", 
    "X-Amzn-Trace-Id": "Root=1-65996b0f-40682f311eeff88a35493058"
  }

response = requests.get(quote_toscrape_web_page)

soup1 = BeautifulSoup(response.content, 'html.parser')

web_content_parsed = BeautifulSoup(soup1.prettify(), 'html.parser')

print(web_content_parsed)
```

```python
web_content_parsed = web_content_parsed.find_all("div", class_="quote")
web_content_parsed
```

```python
quote_author_collection = {

}

for text_within_class in web_content_parsed:

    quote = text_within_class.find("span", class_="text").get_text(strip=True)
    author = text_within_class.find("small", class_="author").get_text(strip=True)
    quote_author_collection[author] = quote

quote_author_collection
```

```python
quote_author_collection.items()
```

```python
df_quote_collection = pd.DataFrame(data=quote_author_collection.items(), columns=["Author", "Quote"])
df_quote_collection.to_csv("success_quote.csv")
```
